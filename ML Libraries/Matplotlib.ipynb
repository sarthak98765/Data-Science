```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Simple Plot

```python
# initializing the data 
x = [10, 20, 30, 40] 
y = [20, 30, 40, 50] 
  
# plotting the data 
plt.plot(x, y) 
  
# Adding the title 
plt.title("Simple Plot") 
  
# Adding the labels 
plt.ylabel("y-axis") 
plt.xlabel("x-axis") 
plt.show()
```

## SubPlot

```python
# data to display on plots  
x = [3, 1, 3]  
y = [3, 2, 1]  
z = [1, 3, 1]  
  
# Creating figure object  
plt.figure()  
  
# adding first subplot  
plt.subplot(121)  
plt.plot(x, y)  
  
# adding second subplot  
plt.subplot(122)  
plt.plot(z, y)
```

## Legend

```python
# data to display on plots  
x = [3, 1, 3]  
y = [3, 2, 1]  
plt.plot(x, y) 
plt.plot(y, x) 
  
# Adding the legends 
plt.legend(["blue", "red"]) 
plt.show()
```

## Line Chart

```python
# data to display on plots  
x = [3, 1, 3]  
y = [3, 2, 1]  
  
# This will plot a simple line chart 
# with elements of x as x axis and y 
# as y axis 
plt.plot(x, y) 
plt.title("Line Chart") 
  
# Adding the legends 
plt.legend(["Line"]) 
plt.show()
```
