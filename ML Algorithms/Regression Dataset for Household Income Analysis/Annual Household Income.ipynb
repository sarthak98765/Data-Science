## Dataset Description

This dataset contains demographic and socio-economic information about households. The features included are:

- **Age**: Age of the primary household member (range: 18 to 70 years).
- **Education Level**: Highest education level attained by the primary household member. Categories include:
  - High School
  - Bachelor's
  - Master's
  - Doctorate
- **Occupation**: Type of occupation of the primary household member. Categories include:
  - Healthcare
  - Education
  - Technology
  - Finance
  - Others
- **Number of Dependents**: Number of dependents in the household (range: 0 to 5).
- **Location**: Residential location of the household. Categories include:
  - Urban
  - Suburban
  - Rural
- **Work Experience**: Years of work experience of the primary household member (range: 0 to 50 years).
- **Marital Status**: Marital status of the primary household member. Categories include:
  - Single
  - Married
  - Divorced
- **Employment Status**: Employment status of the primary household member. Categories include:
  - Full-time
  - Part-time
  - Self-employed
- **Household Size**: Total number of individuals living in the household (range: 1 to 7).
- **Homeownership Status**: Homeownership status of the household. Categories include:
  - Own
  - Rent
- **Type of Housing**: Type of housing the household resides in. Categories include:
  - Apartment
  - Single-family home
  - Townhouse
- **Gender**: Gender of the primary household member. Categories include:
  - Male
  - Female
- **Primary Mode of Transportation**: Primary mode of transportation used by the household member. Categories include:
  - Car
  - Public transit
  - Biking
  - Walking
- **Annual Household Income**: Actual annual household income, derived from a combination of features with added noise. Unit: USD

# 4. Exploratory Data Analysis (EDA)

### Key Features of EDA

#### 1. Understanding the Data
- EDA helps in understanding the structure, distribution, and characteristics of the dataset.
- Provides insight into the nature of the data, such as its range, variability, and central tendencies.
- Common techniques:
  - Descriptive statistics (mean, median, mode, standard deviation, etc.)
  - Data distribution visualizations (histograms, box plots, etc.)

#### 2. Identifying Patterns and Relationships
- EDA techniques allow for the identification of patterns, trends, and relationships within the data.
- Discover correlations between variables, detect outliers, and uncover hidden patterns.
- Common techniques:
  - Scatter plots, correlation matrices, heatmaps, pair plots

#### 3. Feature Selection
- EDA helps in identifying important features or variables that are relevant for modeling or analysis.
- Examining the relationships between variables and their associations with the target variable.
- Common techniques:
  - Feature correlation analysis, importance ranking, variance analysis

#### 4. Data Cleaning and Preprocessing
- EDA helps in detecting and addressing data quality issues such as missing values, outliers, and inconsistencies.
- Identifying problematic data points early in the analysis process to take appropriate steps to clean and preprocess the data.
- Common techniques:
  - Missing value analysis, outlier detection, data transformation

# 5. Feature Encoding

```python
from sklearn.model_selection import KFold
target = 'Income'
df_encoded = df.copy()

kf = KFold(n_splits=5, shuffle=True, random_state=1)

for col in cat_cols:
    col_encoded = f'{col}_encoded'
    df_encoded[col_encoded] = 0
    
    for train_index, val_index in kf.split(df):
        X_train, X_val = df.iloc[train_index], df.iloc[val_index]
        means = X_train.groupby(col)[target].mean()
        df_encoded.loc[val_index, col_encoded] = df.loc[val_index, col].map(means)
    df_encoded[col_encoded].fillna(df[target].mean(), inplace=True)
    df_encoded.drop(columns=[col], inplace=True)

df_encoded.head()
```

# 6. Train-test split

```python
x=df_encoded.drop(["Income"],axis=1)
y=df_encoded["Income"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# 7. Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns, index=x_train.index)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns, index=x_test.index)

x_train_scaled.head()
```

# 8. Model Creation

## 8.1 Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
```

## 8.2. Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
gbr.fit(x_train, y_train)
y_pred_gbr = gbr.predict(x_test)
```

## 8.3 Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet
en = ElasticNet()
en.fit(x_train, y_train)
y_pred_en = en.predict(x_test)
```

## 8.4. Lasso Regression

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(x_train, y_train)
y_pred_lasso = lasso.predict(x_test)
```

## 8.5 Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
y_pred_dt = dt.predict(x_test)
```

## 8.6. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
```

# 9. Model Selection

### Comparing All Evaluation Metrics in Increasing Order

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Initialize an empty DataFrame for model comparison
model_comparison = pd.DataFrame(columns=['Model', 'MAE', 'MSE', 'RMSE', 'RÂ²'])

# Define a function to add the model's metrics to the DataFrame
def add_model_metrics(model_name, y_test, y_pred):
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    
    return pd.DataFrame({
        'Model': [model_name],
        'MAE': [mae],
        'MSE': [mse],
        'RMSE': [rmse],
        'RÂ²': [r2]
    })

# Gradient Boosting Regressor
model_comparison = pd.concat([model_comparison, add_model_metrics('Gradient Boosting Regressor', y_test, y_pred_gbr)], ignore_index=True)

# Linear Regression
model_comparison = pd.concat([model_comparison, add_model_metrics('Linear Regression', y_test, y_pred_lr)], ignore_index=True)

# Elastic Net Regression
model_comparison = pd.concat([model_comparison, add_model_metrics('Elastic Net Regression', y_test, y_pred_en)], ignore_index=True)

# Lasso Regression
model_comparison = pd.concat([model_comparison, add_model_metrics('Lasso Regression', y_test, y_pred_lasso)], ignore_index=True)

# Decision Tree Regressor
model_comparison = pd.concat([model_comparison, add_model_metrics('Decision Tree Regressor', y_test, y_pred_dt)], ignore_index=True)

# Random Forest Regressor
model_comparison = pd.concat([model_comparison, add_model_metrics('Random Forest Regressor', y_test, y_pred_rf)], ignore_index=True)

# Print the model comparison table
print(model_comparison)

# Plotting all evaluation metrics in increasing order
metrics = ['MAE', 'MSE', 'RMSE', 'RÂ²']
plt.figure(figsize=(14, 10))

for i, metric in enumerate(metrics, 1):
    sorted_comparison = model_comparison.sort_values(by=metric, ascending=(metric != 'RÂ²'))
    plt.subplot(2, 2, i)
    plt.bar(sorted_comparison['Model'], sorted_comparison[metric])
    plt.xlabel('Model')
    plt.ylabel(metric)
    plt.title(f'Model Comparison: {metric}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
