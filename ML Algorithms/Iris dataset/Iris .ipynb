# 4. Exploratory Data Analysis (EDA)

### Key Features of EDA

#### 1. Understanding the Data
- EDA helps in understanding the structure, distribution, and characteristics of the dataset.
- Provides insight into the nature of the data, such as its range, variability, and central tendencies.
- Common techniques:
  - Descriptive statistics (mean, median, mode, standard deviation, etc.)
  - Data distribution visualizations (histograms, box plots, etc.)

#### 2. Identifying Patterns and Relationships
- EDA techniques allow for the identification of patterns, trends, and relationships within the data.
- Discover correlations between variables, detect outliers, and uncover hidden patterns.
- Common techniques:
  - Scatter plots, correlation matrices, heatmaps, pair plots

#### 3. Generating Hypotheses
- EDA often leads to the generation of hypotheses or research questions.
- By exploring the data visually and quantitatively, analysts can formulate hypotheses about the underlying factors driving observed patterns.
- Common techniques:
  - Visual exploration, trend analysis, subgroup analysis

#### 4. Feature Selection
- EDA helps in identifying important features or variables that are relevant for modeling or analysis.
- Examining the relationships between variables and their associations with the target variable.
- Common techniques:
  - Feature correlation analysis, importance ranking, variance analysis

#### 5. Data Cleaning and Preprocessing
- EDA helps in detecting and addressing data quality issues such as missing values, outliers, and inconsistencies.
- Identifying problematic data points early in the analysis process to take appropriate steps to clean and preprocess the data.
- Common techniques:
  - Missing value analysis, outlier detection, data transformation

#### 6. Communicating Results
- EDA provides visualizations and summaries that facilitate communication of results to stakeholders.
- Visualizations such as charts, graphs, and summary statistics make it easier to convey complex findings and insights in a clear and understandable manner.
- Common techniques:
  - Bar charts, pie charts, line graphs, summary tables

#### 7. Informing Decision-Making
- EDA helps in making informed decisions by providing insights into the data that can guide subsequent analysis, modeling, and decision-making processes.
- Understanding the underlying structure of the data allows for more accurate predictions and effective strategies.
- Common techniques:
  - Predictive modeling, scenario analysis, strategic planningrategies

#### 4.1.2 Boxplot

A boxplot, also known as a box-and-whisker plot, provides a visual summary of the distribution of a dataset. It displays several descriptive statistics such as the median, quartiles, and potential outliers.

##### Key Elements of a Boxplot

1. **Median (Q2/50th Percentile)**: The middle value of the dataset. It represents the point where half of the data is below and half is above.
2. **Quartiles (Q1, Q3)**: The dataset is divided into four equal parts. Q1 (lower quartile) is the value below which 25% of the data falls, and Q3 (upper quartile) is the value below which 75% of the data falls.
3. **Interquartile Range (IQR)**: The range between the first and third quartiles (Q3 - Q1). It describes the middle 50% of the data.
4. **Whiskers**: Lines extending from the box that represent the range of the data, excluding outliers. The whiskers typically extend to 1.5 times the IQR from the quartiles.
5. **Outliers**: Data points that fall outside the whiskers. These points are typically plotted individually.
6. **Box Length**: The length of the box represents the spread of the middle 50% of the data. A longer box indicates a larger spread.

###### Interpretation of a Boxplot

By observing a boxplot, you can quickly grasp key aspects of the distribution of your data, such as its central tendency, spread, and presence of outliers. It's particularly useful for comparing distributions between different groups or datasets.

- **Central Tendency**: The median line within the box.
- **Spread**: The length of the box (IQR) and the range covered by the whiskers.
- **Outliers**: Points outside the whiskers.
- **Symmetry**: If the median is centered in the box and whiskers are of similar length on both sides, the data may be symmetric.
- **Skewness**: If the median is not centered and whiskers are of different lengths, the data may be skewed.

A pairplot is a visualization technique used to plot pairwise relationships in a dataset. It is particularly useful for exploring the interactions between different numerical features.

Key Features:
- Scatter Plots: For each pair of numerical features, a scatter plot is created to visualize their relationship.
- Histograms/Kernel Density Plots: For each numerical feature, a histogram or KDE plot is displayed along the diagonal to show its distribution.
- Hue: An optional parameter that allows for coloring the data points by a categorical feature, facilitating comparison between groups.

### 4.2.2 Countplot
A countplot is a visualization technique used to show the frequency distribution of categorical data. It is essentially a bar plot where the height of each bar corresponds to the number of instances of each category.

Key Features:
- Counts: The primary focus is on the count of occurrences of each category.
- Hue: An optional parameter to further divide the bars based on another categorical feature, allowing for grouped comparisons.
- Orientation: Can be displayed horizontally or vertically depending on the data and preference.

```python
for col in df.columns:
    if col != "Species":
        plt.figure(figsize=(12,4))
        sns.countplot(x=col, hue="Species", data=df)
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.title(f"{col} vs Species")
        plt.xticks(rotation=45) 
        plt.show()
```

## 5.1 Removing Outliers

```python
# IQR
Q1 = np.percentile(df['SepalWidthCm'], 25, method='midpoint')
print("Q1 is:",Q1)
Q3 = np.percentile(df['SepalWidthCm'], 75, method='midpoint')
print("Q3 is:",Q3)
IQR = Q3 - Q1
print("IQR is:",IQR)
```

```python
upper = Q3 +1.5*IQR
lower = Q1 - 1.5*IQR
```

```python
# Above Upper bound
upper_array = np.array(df['SepalWidthCm'] >= upper)
print("Upper Bound:", upper)
print(upper_array.sum())
 
# Below Lower bound
lower_array = np.array(df['SepalWidthCm'] <= lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
df.isnull().sum()
```

```python
print("Old Shape: ", df.shape)
# Removing the outliers
# Remove outliers and assign the modified DataFrame to df3
df3 = df[~(upper_array | lower_array)]

# Print the new shape of the DataFrame
print("New Shape:", df3.shape)
```

# 6. Train-test split

```python
x=df3.drop(["Species"],axis=1)
y=df3["Species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# 7. Model Creation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

## 7.1 Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
y_pred_lr=LR.predict(x_test)
```

## 7.2 Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred_dt = dt.predict(x_test)
```

## 7.3 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
```

## 7.4 Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
y_pred_nb = nb.predict(x_test)
```

## 7.5 K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
```

# 8. Model Selection

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import matplotlib.pyplot as plt

# Initialize an empty DataFrame for model comparison
model_comparison = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

# Logistic Regression
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr, average='weighted')
recall_lr = recall_score(y_test, y_pred_lr, average='weighted')
f1_lr = f1_score(y_test, y_pred_lr, average='weighted')

model_comparison = pd.concat([model_comparison, pd.DataFrame({
    'Model': ['Logistic Regression'],
    'Accuracy': [accuracy_lr],
    'Precision': [precision_lr],
    'Recall': [recall_lr],
    'F1 Score': [f1_lr]
})], ignore_index=True)

# Decision Trees
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt, average='weighted')
recall_dt = recall_score(y_test, y_pred_dt, average='weighted')
f1_dt = f1_score(y_test, y_pred_dt, average='weighted')

model_comparison = pd.concat([model_comparison, pd.DataFrame({
    'Model': ['Decision Trees'],
    'Accuracy': [accuracy_dt],
    'Precision': [precision_dt],
    'Recall': [recall_dt],
    'F1 Score': [f1_dt]
})], ignore_index=True)

# Random Forest
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf, average='weighted')
recall_rf = recall_score(y_test, y_pred_rf, average='weighted')
f1_rf = f1_score(y_test, y_pred_rf, average='weighted')

model_comparison = pd.concat([model_comparison, pd.DataFrame({
    'Model': ['Random Forest'],
    'Accuracy': [accuracy_rf],
    'Precision': [precision_rf],
    'Recall': [recall_rf],
    'F1 Score': [f1_rf]
})], ignore_index=True)

# Naive Bayes
accuracy_nb = accuracy_score(y_test, y_pred_nb)
precision_nb = precision_score(y_test, y_pred_nb, average='weighted')
recall_nb = recall_score(y_test, y_pred_nb, average='weighted')
f1_nb = f1_score(y_test, y_pred_nb, average='weighted')

model_comparison = pd.concat([model_comparison, pd.DataFrame({
    'Model': ['Naive Bayes'],
    'Accuracy': [accuracy_nb],
    'Precision': [precision_nb],
    'Recall': [recall_nb],
    'F1 Score': [f1_nb]
})], ignore_index=True)

# K-Nearest Neighbors (KNN)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn, average='weighted')
recall_knn = recall_score(y_test, y_pred_knn, average='weighted')
f1_knn = f1_score(y_test, y_pred_knn, average='weighted')

model_comparison = pd.concat([model_comparison, pd.DataFrame({
    'Model': ['K-Nearest Neighbors (KNN)'],
    'Accuracy': [accuracy_knn],
    'Precision': [precision_knn],
    'Recall': [recall_knn],
    'F1 Score': [f1_knn]
})], ignore_index=True)

# Sort the DataFrame by Accuracy
model_comparison = model_comparison.sort_values(by='Accuracy', ascending=False)

# Print the model comparison table
print(model_comparison)

# Bar plot of Accuracy scores
plt.figure(figsize=(10, 6))
plt.bar(model_comparison['Model'], model_comparison['Accuracy'])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Comparison: Accuracy')
plt.xticks(rotation=45)
plt.show()
```
