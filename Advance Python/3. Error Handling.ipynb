## Difference Between Errors and Exceptions
Errors and exceptions are both issues in a program, but they differ in severity and handling. Let's see how:

- Error: Serious problems in the program logic that cannot be handled. Examples include syntax errors or memory errors.
- Exception: Less severe problems that occur at runtime and can be managed using exception handling (e.g., invalid input, missing files).

````python
# Syntax Error (Error)
print("Hello world"  # Missing closing parenthesis
````

````python
# ZeroDivisionError (Exception)
n = 10
res = n / 0
````

A syntax error stops the code from running at all, while an exception like ZeroDivisionError occurs during execution and can be caught with exception handling.

Python provides four main keywords for handling exceptions: try, except, else and finally each plays a unique role. Let's see syntax:

1. try: Runs the risky code that might cause an error.
2. except: Catches and handles the error if one occurs.
3. else: Executes only if no exception occurs in try.
4. finally: Runs regardless of what happens useful for cleanup tasks like closing files.

````python
def withdraw_money(balance, amount):
    try:
        # Risky operation: user may enter invalid amount
        amount = float(amount)

        if amount > balance:
            raise ValueError("Insufficient balance")

    except ValueError as e:
        # Handles invalid input or custom raised error
        print(f"Transaction failed: {e}")

    else:
        # Runs only if no exception occurred
        balance -= amount
        print(f"Withdrawal successful. Remaining balance: â‚¹{balance}")

    finally:
        # Always runs (success or failure)
        print("Transaction session ended.\n")


# Test cases
withdraw_money(5000, "2000")   # Valid
withdraw_money(5000, "7000")   # Insufficient balance
withdraw_money(5000, "abc")    # Invalid input
````

### 1. TypeError
Occurs when an operation is applied to an incompatible data type.

Example scenarios:
- Adding string + integer
- Passing wrong type to a function
- Performing numeric ops on non-numeric data

````python
def add_numbers(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both inputs must be numbers")
    return a + b
````

````python
add_numbers(7,9.5)
````

````python
add_numbers(7,"9.5")
````

### 2. ValueError
Occurs when the type is correct but the value is invalid.

Example scenarios:
- Negative age
- Invalid date
- String that canâ€™t convert to int

````python
def withdraw(balance, amount):
    if amount <= 0:
        raise ValueError("Amount must be positive")
````

````python
withdraw(500,100)    
````

````python
withdraw(500,-100)
````

### 3. IndexError
Raised when accessing a list index that doesnâ€™t exist.

````python
def get_item(lst, index):
    if index >= len(lst):
        raise IndexError("Index out of range")
    return lst[index]
````

````python
get_item([1,2,3], 1)   
````

````python
get_item([1,2,3], 5)   
````

### 4. KeyError
Occurs when a dictionary key is missing.

````python
def get_user(data):
    if "name" not in data:
        raise KeyError("User name missing")
    return data["name"]
````

````python
get_user({"name": "Sarthak"}) 
````

````python
get_user({"id": 1})            
````

### 5. FileNotFoundError
Raised when file operations fail due to missing file.

````python
import os

def read_file(path):
    if not os.path.exists(path):
        raise FileNotFoundError("File does not exist")
````

````python
read_file("data.txt") 
````

## All exceptions in 1 example

````python
def system_validator(age, amount, numbers, user_data, file_path):
    try:
        if not isinstance(age, int):
            raise TypeError("Age must be integer")

        if age < 0:
            raise ValueError("Age cannot be negative")

        if amount <= 0:
            raise ValueError("Amount must be positive")

        if len(numbers) < 2:
            raise IndexError("List must contain at least 2 items")

        if "name" not in user_data:
            raise KeyError("User name missing")

        import os
        if not os.path.exists(file_path):
            raise FileNotFoundError("File path invalid")

    except Exception as e:
        print(f"Error occurred: {e}")

    else:
        print("All validations passed âœ…")

    finally:
        print("Validation process completed.\n")
````

### âœ… Test Case 1 â€” All Valid Inputs
