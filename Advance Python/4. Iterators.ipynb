## Iterable Protocol
An **Iterable** is any object that can return its elements one by one.

It does **not** store iteration logic itself â€” it only knows how to create an iterator.

Python checks if an object is iterable by calling:

```python
obj.__iter__()
```

If this method exists â†’ object is iterable.

Common Iterables:
- List
- Tuple
- String
- Dictionary
- Set
- File objects

So, When we run:

```Python
for x in my_list:
    pass
```
Python internally performs the following steps:
- Calls iter(my_list)
- Gets the iterator
- Fetches values using next()

````python
my_list = [10, 20, 30]

# Convert iterable â†’ iterator
iterator = iter(my_list)

print(next(iterator))  # 10
print(next(iterator))  # 20
print(next(iterator))  # 30
````

Raises StopIteration.                   
So,
1. Iterable = container of data
2. Has __iter__()
3. Does NOT track position
4. Can create multiple iterators

## Iterator Protocol

An **Iterator** is an object that:
- Keeps track of current position  
- Returns next value  
- Knows when data ends  

An Iterator must implement:

```python
__iter__()
__next__()
```
Flow: 
1. __iter__() â†’ returns itself
2. __next__() â†’ returns next value
3. Raises StopIteration when done

````python
my_list = [1, 2, 3]

iterator = iter(my_list)

while True:
    try:
        item = next(iterator)
        print(item)
    except StopIteration:
        break
````

This is exactly how for loop works internally.

So,
1. Iterator = data + state
2. Moves forward only
3. Cannot reset automatically
4. Memory efficient

## 4.3 __iter__ and __next__

These are dunder methods that implement iterator behavior.
- __iter__() ->	Returns iterator object
- __next__() ->	Returns next value

````python
class Counter:
    def __init__(self,limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.limit:
            self.current += 1
            return self.current
        else:
            raise StopIteration
````

````python
counter = Counter(5)

for num in counter:
    print(num)
````

### Mini Real-World Program â€” Log File Processor

#### Problem

A company wants to scan a **server log file** and extract only `"ERROR"` lines.

Log file is huge â†’ cannot load fully.

So we use an iterator.

## ðŸ“„ Sample Log File (`server.log`)

- INFO User logged in
- INFO Payment processed
- ERROR Database connection failed
- INFO Email sent
- ERROR Timeout occurred

````python
class ErrorLogIterator:
    def __init__(self, filepath):
        self.file = open(filepath, 'r')

    def __iter__(self):
        return self

    def __next__(self):
        while True:
            line = self.file.readline()

            if not line:
                self.file.close()
                raise StopIteration

            if "ERROR" in line:
                return line.strip()
````

````python
logs = ErrorLogIterator("log_file/server.log")

for error in logs:
    print(error)
````
